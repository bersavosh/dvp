# Seaborn palettes

```python
# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()
```

```python
# Load an example dataset
tips = sns.load_dataset("tips")
tips
```

```python
fig = sns.relplot(data=tips,
                  x="total_bill", y="tip", col="time",
                  hue="smoker", style="smoker", size="size")
```

## Palette

We can inspect and use available color palettes with `color_palette`.

Calling with no arguments returns all colors from the current default color cycle:

```python
sns.color_palette()
```

Possible `palette` values include:
### Name of a seaborn palette 

(`deep`, `muted`, `bright`, `pastel`, `dark`, `colorblind`)

```python
for i in ['deep', 'muted', 'bright', 'pastel', 'dark', 'colorblind']:
    print(i)
    palette = sns.color_palette(i)
    display(palette)
```

### Name of known matplotlib, seaborn, HUSL, Color Brewer colormaps

(*to return a specified number of evenly spaced colors*)

```python
for i in ['flare','viridis','bwr','husl','hls','Set1']:
    print(i,'(discrete)')
    palette = sns.color_palette(i,n_colors=10)
    display(palette)
    print(i,'(continuous)')
    palette = sns.color_palette(i,as_cmap=True)
    display(palette)
```

As you can see, `Set1` is a qualitative (discrete) color palette, and thus even as a colormap it appears as such.

### a customized cubehelix color palette 

`ch:<cubehelix arguments>`

```python
print('(discrete)')
palette = sns.color_palette("ch:s=.25,rot=-.25",n_colors=10)
display(palette)
print('(continuous)')
palette = sns.color_palette("ch:s=.25,rot=-.25", as_cmap=True)
display(palette)

```

### a light-themed sequential colormap to a seed color

```python
print('(discrete)')
palette = sns.color_palette("light:#5A9",n_colors=10)
display(palette)
print('(continuous)')
palette = sns.color_palette("light:#5A9", as_cmap=True)
display(palette)

```

#### How do we use these for visualization?

- For seaborn defaults, we can just pass the name. 
- For ones we have made, we can pass the produced `sns.color_palette` to `set_palette` or `set_theme(palette='...')`

```python
palt = sns.color_palette("pastel")
display(palt)
sns.set_palette(palt)

# Above is equivalent of:
#sns.set_theme(palette=palt)

fig = sns.relplot(data=tips,
                  x="total_bill", y="tip", col="time",
                  hue="smoker", style="smoker", size="size")
```

```python
palt = sns.color_palette("hot",n_colors=2,desat=1)
display(palt)
sns.set_palette(palt)

# Above is equivalent of:
#sns.set_theme(palette=palt)

fig = sns.relplot(data=tips,
                  x="total_bill", y="tip", col="time",
                  hue="smoker", style="smoker", size="size")
```
