# Seaborn contexts

```python
# Import seaborn
import seaborn as sns
import matplotlib.pyplot as plt

# Apply the default theme
sns.set_theme()
```

```python
# Load an example dataset
tips = sns.load_dataset("tips")
tips
```

```python
fig = sns.relplot(data=tips,
                  x="total_bill", y="tip", col="time",
                  hue="smoker", style="smoker", size="size")
```

## Context

There are four general theme contexts to govern visualization elements:

`paper`, `notebook`, `talk`, `poster`

```python
sns.set_theme()
for context in ['paper','notebook','talk','poster']:
    sns.set_context(context)
    # Above is equivalent of:
    #sns.set_theme(context=context)

    fig = sns.relplot(data=tips,
                      x="total_bill", y="tip", col="time",
                      hue="smoker", style="smoker", size="size")
    
    ## The following bit is just to force organized print and show commands
    plt.close(fig.figure)
    print(context)
    display(fig.figure)
```

## context font scale

```python
sns.set_theme()

for fsc in [0.5,1.0,1.5]:
    sns.set_context('notebook', font_scale=fsc)
    # Above is equivalent of:
    #sns.set_theme(context='notebook',font_scale=fsc)

    fig = sns.relplot(data=tips,
                      x="total_bill", y="tip", col="time",
                      hue="smoker", style="smoker", size="size")
    
    ## The following bit is just to force organized print and show commands
    plt.close(fig.figure)
    print('font scale =',fsc)
    display(fig.figure)
```
