# Seaborn and matplotlib fonts

```python
# Import seaborn
import seaborn as sns
import matplotlib
from IPython.core.display import HTML
```

```python
# Load an example dataset
tips = sns.load_dataset("tips")
tips
```

## Fonts in a theme

Default is sans-serif font

```python
sns.set_theme(font='sans-serif')
# above is equivalent to
#sns.set_theme()

fig = sns.relplot(data=tips,
                  x="total_bill", y="tip", col="time",
                  hue="smoker", style="smoker", size="size")
```

```python
sns.set_theme(font='serif')
# above is equivalent to
#sns.set_theme()

fig = sns.relplot(data=tips,
                  x="total_bill", y="tip", col="time",
                  hue="smoker", style="smoker", size="size")
```

```python
sns.set_theme(font='purisa')

fig = sns.relplot(data=tips,
                  x="total_bill", y="tip", col="time",
                  hue="smoker", style="smoker", size="size")
```

## Here is a fancy way of seeing a list of all available fonts to matplotlib on a system:

```python
def html_display(fontname):
    return f"<p>{fontname}: <span style='font-family:{fontname}; font-size: 24px;'>{fontname}</p>"

fontobject_list = matplotlib.font_manager.fontManager.ttflist

fontname_list = sorted(set([fontobject.name for fontobject in fontobject_list]))


code = "\n".join([html_display(font) for font in fontname_list])

HTML("<div style='column-count: 3;'>{}</div>".format(code))
```
