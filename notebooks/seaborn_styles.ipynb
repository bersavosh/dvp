# Seaborn styles

```python
# Import seaborn
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Apply the default theme
sns.set_theme()
```

```python
# Load an example dataset
tips = sns.load_dataset("tips")
tips
```

```python
fig = sns.relplot(data=tips,
                  x="total_bill", y="tip", col="time",
                  hue="smoker", style="smoker", size="size")
```

## Styles

it can be one of the following five:

- `darkgrid`
- `dark`
- `whitegrid`
- `white`
- `ticks`

```python
for style in ['darkgrid','dark','whitegrid','white','ticks']:
    sns.set_style(style)
    # Above is equivalent of:
    #sns.set_theme(style='white')

    fig = sns.relplot(data=tips,
                      x="total_bill", y="tip", col="time",
                      hue="smoker", style="smoker", size="size")
    
    ## The following bit is just to force organized print and show commands
    plt.close(fig.figure)
    print(style)
    display(fig.figure)
```
