# Cartographic projections in matplotlib

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = plt.subplot(projection="aitoff")
ax.set_title("Aitoff")
ax.grid(True)
```

```python
fig = plt.figure()
ax = plt.subplot(projection="hammer")
ax.set_title("Hammer")
ax.grid(True)
```

```python
fig = plt.figure()
ax = plt.subplot(projection="lambert")
ax.set_title("Lambert")
ax.grid(True)
```
